## 1. Extraindo texto de uma **Imagem**

```python
from google.colab import files
from google.colab.patches import cv2_imshow
import cv2

# Upload de Arquivos
uploaded = files.upload()

# Carregar imagem
imagem = cv2.imread(list(uploaded.keys())[0])

# Exibir imagem
cv2_imshow(imagem)
```

```python
# Carregando imagem sem Botão de Upload
img = cv2.imread('/content/imagem_com_texto.png')

# Exibir imagem
cv2_imshow(img)
```

```python
# Aplicando o Tesseract OCR
!pip install pytesseract
# Instala o motor do Tesseract e o suporte para a língua portuguesa
!apt-get install tesseract-ocr
!apt-get install tesseract-ocr-por
import pytesseract

texto = pytesseract.image_to_string(img, lang='por')
print("\nTexto Extraído: ")
print(texto)
```

```python
# Convertendo a imagem para grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Exibir a imagem grayscale
cv2_imshow(img_gray)
```

```python
# Aplicando o Tesseract OCR
texto = pytesseract.image_to_string(img_gray, lang='por')
print("Texto Extraído da imagem em Grayscale: \n\n" + texto)
```

```python
# Redimensionando a Imagem para padronizar o tamanho das figuras de entrada
img_redimensionada = cv2.resize(img, (400, 100))

# Exibir imagem redimensionada
cv2_imshow(img_redimensionada)
```

```python
# Aplicando o Tesseract OCR
texto = pytesseract.image_to_string(img_redimensionada, lang='por')
print("Texto Extraído da imagem em redimensionada: \n\n" + texto)
```

```python
# Aplicar filtro de suavização para reduzir o ruido e melhorar a detecção de bordas
img_suavizada = cv2.GaussianBlur(img, (15, 15), 0)

# Exibir imagem suavizada
cv2_imshow(img_suavizada)
```

```python
# Aplicando o Tesseract OCR
texto = pytesseract.image_to_string(img_suavizada, lang='por')
print("Texto Extraído da imagem em redimensionada: \n\n" + texto)
```

```python
# Detecção de bordas com Canny
bordas = cv2.Canny(img_gray, 100, 200)

# Exibir Bordas Detectadas
cv2_imshow(bordas)
```

```python
# Aplicando o Tesseract OCR
texto = pytesseract.image_to_string(bordas, lang='por')
print("Texto Extraído das bordas detectadas na imagem: \n\n" + texto)
```
